```python
# keep

import torch
import os
import sys

model_version = slai.model_version("{{SLAI_PROJECT_NAME}}/{{SLAI_MODEL_NAME}}/initial")
```

```python
# keep

# Create a model
import torch.nn as nn
import torch.nn.functional as F


class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(2, 120)
        self.fc2 = nn.Linear(120, 40)
        self.fc3 = nn.Linear(40, 10)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return x


model = Net()
```
